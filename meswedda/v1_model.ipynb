```python
import pickle
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# ── 1. Load Processed Data ───────────────────────────────────────
print("Loading data artifacts...")
with open('../data/X_train.pkl', 'rb') as f:
    X_train = pickle.load(f)
with open('../data/X_test.pkl', 'rb') as f:
    X_test = pickle.load(f)
with open('../data/y_train.pkl', 'rb') as f:
    y_train = pickle.load(f)
with open('../data/y_test.pkl', 'rb') as f:
    y_test = pickle.load(f)
```

```python
# ── 2. Helper Function for Evaluation ────────────────────────────
def test_model(name, model):
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)
    mae_pct = (np.exp(mae) - 1) * 100  # Back-transform log-scale error
    
    # Print results cleanly
    print(f"--- {name} ---")
    print(f"  R²  : {r2:.4f}")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  MAE : {mae:.4f} (±{mae_pct:.1f}%)")
    print("-" * 35)
```

```python
print("\nTraining and evaluating models...\n")

# Ordinary Least Squares (Standard Linear Regression)
# Note: Might perform poorly or overfit due to the 370+ One-Hot Encoded columns
lr_model = LinearRegression()
test_model("Linear Regression (OLS)", lr_model)
```

```python
# Ridge Regression (L2 Penalty)
# The baseline you currently use in your main pipeline
ridge_model = Ridge(alpha=1.0)
test_model("Ridge Regression (alpha=1.0)", ridge_model)
```

```python
# Lasso Regression (L1 Penalty)
# Note: We use a small alpha (0.001) because Lasso aggressively drops features to 0
lasso_model = Lasso(alpha=0.001) 
test_model("Lasso Regression (alpha=0.001)", lasso_model)
```
